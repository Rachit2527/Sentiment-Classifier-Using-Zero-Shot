```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/india-news-headlines.csv')
```

```python
data
```

```python
samples = data.sample(100)
```

```python
strings = samples.headline_text.values
```

```python
strings
```

```python
sequence = list(strings)
```

```python
sequence
```

```python
labels = ['politics','finance','sports','entertainment','agriculture']
```

```python
from transformers import pipeline
classifier = pipeline('zero-shot-classification' , device = 0)
```

```python
output = classifier(sequence,labels)
```
